```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# Initialize WebDriver
driver = webdriver.Chrome()
driver.get("https://www.imdb.com/search/title/?title_type=feature&release_date=2024-01-01,2024-12-31&genres=action")
driver.maximize_window()

# Wait for the page to load
time.sleep(5)

# Click the "View More" button multiple times to load more data
for _ in range(9):
    try:
        view_more_button = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, "//button[contains(@class, 'ipc-see-more__button')]"))
        )
        view_more_button.send_keys(Keys.ENTER)  # Press ENTER key to click
        time.sleep(3)  # Wait for new content to load
    except Exception as e:
        print("No more 'View More' button found or an error occurred:", e)
        break

# Extract movie titles using h3 tag and class ipc-title__text
action_movie_titles_list = []
for title in driver.find_elements(By.XPATH, "//h3[contains(@class, 'ipc-title__text')]"):
    if title.text.strip():  # Ensure it's not empty
        action_movie_titles_list.append(title.text)

# Print movie titles (limit to 500)
for title in action_movie_titles_list[:500]:
    print(title)

# Quit WebDriver
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import re
import time

# Initialize WebDriver
driver = webdriver.Chrome()
driver.get("https://www.imdb.com/search/title/?title_type=feature&release_date=2024-01-01,2024-12-31&genres=action")
driver.maximize_window()

# Wait for the page to load
time.sleep(5)

# Click the "View More" button multiple times to load more data
for _ in range(9):
    try:
        view_more_button = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, "//button[contains(@class, 'ipc-see-more__button')]"))
        )
        view_more_button.send_keys(Keys.ENTER)  # Press ENTER key to click
        time.sleep(3)  # Wait for new content to load
    except Exception as e:
        print("No more 'View More' button found or an error occurred:", e)
        break

# Extract movie titles and clean them
raw_titles = [title.text for title in driver.find_elements(By.XPATH, "//h3[contains(@class, 'ipc-title__text')]")]
titles = [re.sub(r'^\d+\.\s*', '', title) for title in raw_titles]  # Remove index numbers (e.g., "1. Movie Name" → "Movie Name")

# Extract ratings, durations, and votes
ratings_elements = driver.find_elements(By.XPATH, "//span[contains(@class, 'ipc-rating-star--rating')]")
durations_elements = driver.find_elements(By.XPATH, "//span[contains(@class, 'sc-f30335b4-7')]")
votes_elements = driver.find_elements(By.XPATH, "//span[contains(@class, 'ipc-rating-star--voteCount')]")

# Ensure 1:1 mapping by adding "N/A" where data is missing
ratings = []
durations = []
votes = []

for i in range(len(titles)):
    # Get rating
    ratings.append(ratings_elements[i].text if i < len(ratings_elements) and ratings_elements[i].text.strip() else "N/A")

    # Get duration and filter correct format (contains 'h' or 'm')
    duration_text = durations_elements[i].text if i < len(durations_elements) else "N/A"
    duration_match = re.search(r'(\d+\s*h\s*\d*\s*m*|\d+\s*m|\d+\s*h)', duration_text)  # Regex to match '1h 30m' or '120m'
    durations.append(duration_match.group() if duration_match else "N/A")

    # Get votes
    votes.append(votes_elements[i].text if i < len(votes_elements) and votes_elements[i].text.strip() else "N/A")

# Store data in a dictionary and create DataFrame
data = {
    "Title": titles[:500],   # Limiting to 500
    "Rating": ratings[:500],
    "Duration": durations[:500],  # Fixed duration format
    "Votes": votes[:500]
}

df = pd.DataFrame(data)

# Print DataFrame
print(df)

# Quit WebDriver
driver.quit()
```

```python
# Action movies 

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
import time

# Initialize WebDriver
driver = webdriver.Chrome()
driver.get("https://www.imdb.com/search/title/?title_type=feature&genres=action&release_date=2024-01-01,2024-12-31")

time.sleep(3)  # Waiting for the page to load

genre = "Action"

# Scroll Until No More New Data Loads
scrolling = True
movies_per_page = 50  # Assuming each load gives 50 movies
total_movies_needed = 500
current_movies = 0

while scrolling and current_movies < total_movies_needed:
    old_page_source = driver.page_source  # Save old page source
    
    # Scroll down to load more data
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)  # Allow time for new data to load
    
    new_page_source = driver.page_source  # Save new page source

    if new_page_source == old_page_source:
        try:
            # Locate and click the "See More" button if present
            see_more_button = driver.find_element(By.XPATH, "//span[contains(@class, 'ipc-see-more')]")
            ActionChains(driver).move_to_element(see_more_button).click().perform()
            time.sleep(2)  # Wait for new content to load
        except Exception:
            scrolling = False  # Stop scrolling if button isn't found
    else:
        current_movies += movies_per_page  # Increment count

# Extract movie containers
movie_blocks = driver.find_elements(By.CSS_SELECTOR, "li.ipc-metadata-list-summary-item")

movies_list = []

for movie in movie_blocks:
    try:
        title = movie.find_element(By.CSS_SELECTOR, "h3.ipc-title__text").text.strip()
        title = title.split(". ", 1)[-1]  # Removes index number if present
    except:
        title = "N/A"

    try:
        duration_element = movie.find_element(By.XPATH, ".//span[contains(@class, 'dli-title-metadata-item') and (contains(text(),'h') or contains(text(),'m'))]")
        duration = duration_element.text.strip() if duration_element.text.strip() else "N/A"
    except:
        duration = "N/A"

    try:
        rating = movie.find_element(By.CSS_SELECTOR, "span.ipc-rating-star--rating").text.strip()
    except:
        rating = "N/A"

    try:
        voting = movie.find_element(By.CSS_SELECTOR, "span.ipc-rating-star--voteCount").text.strip()
    except:
        voting = "N/A"

    movie_data = {
        "Title": title,
        "Genre": genre,
        "Duration": duration,
        "Rating": rating,
        "Votes": voting,
    }
    movies_list.append(movie_data)


# Print results
for movie in movies_list:
    print(movie)

# Close the driver
driver.quit()
```

```python
import pandas as pd

df=pd.DataFrame(movies_list)
df.to_csv("Action.csv")
```
