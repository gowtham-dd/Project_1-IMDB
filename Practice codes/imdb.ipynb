```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
from selenium.webdriver.common.keys import Keys

driver = webdriver.Chrome()
driver.get("https://www.imdb.com/search/title/?title_type=feature&release_date=2024-01-01,2024-12-31")
driver.maximize_window()
time.sleep(5)

scrolling = True

while scrolling:
    old_page_source = driver.page_source
    
    # Get the body element correctly
    body = driver.find_element(By.TAG_NAME, "body")
    body.send_keys(Keys.PAGE_DOWN)  # Use find_element instead of find_elements

    time.sleep(2)  # Allow time for the page to load

    new_page_source = driver.page_source

    if new_page_source == old_page_source:
        scrolling = False

# Extract movie titles
movie_titles = driver.find_elements(By.CLASS_NAME, "ipc-title__text")

# Print movie titles
for title in movie_titles:
    print(title.text)

driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# Initialize WebDriver
driver = webdriver.Chrome()
driver.get("https://www.imdb.com/search/title/?title_type=feature&release_date=2024-01-01,2024-12-31&genres=action")
driver.maximize_window()

# Wait for the page to load
time.sleep(5)

# Click the "View More" button multiple times to load more data
for _ in range(9):
    try:
        view_more_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, "//button[contains(@class, 'ipc-see-more__button')]"))
        )
        driver.execute_script("arguments[0].scrollIntoView();", view_more_button)  # Ensure button is in view
        driver.execute_script("arguments[0].click();", view_more_button)  # Click using JS to avoid interception issues
        time.sleep(3)  # Wait for new content to load
    except Exception as e:
        print("No more 'View More' button found or an error occurred:", e)
        break

# Extract movie titles
action_movie_titles_list = [title.text for title in driver.find_elements(By.CLASS_NAME, "ipc-title__text")]

# Print movie titles (without index)
for title in action_movie_titles_list[:500]:  # Limit to 500 movies
    print(title)

# Quit WebDriver
driver.quit()
```
